```python
import numpy as np
from fractions import Fraction # so that numbers are not displayed in decimal.



#MÃ¡x. x_1 + x_2 st 
#     x_1 + x_2 + x_4 = 8
#     2x_1 + x_2 + x_3 = 10

# Entradas

# A contiene los coeficientes de las constantes de las restricciones
A = np.array([[1, 1, 0, 1], [2, 1, 1, 0]])
# b contiene la cantidad de recursos
b = np.array([8, 10])		
# c contiene los coeficientes  objetivo de la funcion Z	
c = np.array([1, 1, 0, 0])			

# Inicializacion de variables
# B contiene las variables basicas que hacen la matriz identidad
cb = np.array(c[3])
B = np.array([[3], [2]])		

# cb contiene los coeficientes de las variables basicas en la funcion Z

cb = np.vstack((cb, c[2]))	
xb = np.transpose([b])				
# con hstask se combinan la tabla  B y cb
table = np.hstack((B, cb))			
table = np.hstack((table, xb))		
# combine matrices B, cb and xb

# finalmente combina la matriz A para formar la tabla simplex completa
table = np.hstack((table, A))		

# cambia el tipo de tabla a float
table = np.array(table, dtype ='float')

# termina las entradas

# si es un problema de minimizacion, hacer esta variable 1
MIN = 0

print("Table en iteracion = 0")
print("B \tCB \tXB \ty1 \ty2 \ty3 \ty4")
for row in table:
	for el in row:
	
                # limita el denominador a 100
		print(Fraction(str(el)).limit_denominator(100), end ='\t')
	print()
print()
print("Simplex cargando...")


# cuando se alcanza la optimidad, se hara 1
reached = 0	
itr = 1
unbounded = 0
alternate = 0

while reached == 0:

	print("Iteration: ", end =' ')
	print(itr)
	print("B \tCB \tXB \ty1 \ty2 \ty3 \ty4")
	for row in table:
		for el in row:
			print(Fraction(str(el)).limit_denominator(100), end ='\t')
		print()


    # calcula los beneficios relativos -> cj - zj para no basicas
	i = 0
	rel_prof = []
	while i<len(A[0]):
		rel_prof.append(c[i] - np.sum(table[:, 1]*table[:, 3 + i]))
		i = i + 1

	print("beneficio relativo: ", end =" ")
	for profit in rel_prof:
		print(Fraction(str(profit)).limit_denominator(100), end =", ")
	print()
	i = 0
	
	b_var = table[:, 0]

    # comprobando si hay solucion alternativa
	while i<len(A[0]):
		j = 0
		present = 0
		while j<len(b_var):
			if int(b_var[j]) == i:
				present = 1
				break
			j+= 1
		if present == 0:
			if rel_prof[i] == 0:
				alternate = 1
				print("si hay solucion alternativa, se termina el programa")
            
				# print(i, end =" ")
		i+= 1
	print()
	flag = 0
	for profit in rel_prof:
		if profit>0:
			flag = 1
			break
	
        # si todos los beneficios relativos <= 0
	if flag == 0:
		print("todos los beneficios relativos <= 0, optimamente alcanzo")
		reached = 1
		break

	# kth var will enter the basis
    # kth var introduce los basicos
	k = rel_prof.index(max(rel_prof))
	min = 99999
	i = 0
	r = -1
	
    # minimiza la prueba de relacion (solo valores positivos)
	while i<len(table):
		if (table[:, 2][i]>0 and table[:, 3 + k][i]>0):
			val = table[:, 2][i]/table[:, 3 + k][i]
			if val<min:
				min = val
				r = i	 # dejando variable
		i+= 1


        # si no se realizo la prueba de relacion minima
	if r ==-1:
		unbounded = 1
		print("Caso de no acotado")
		break

	print(" elemento pivote indice:", end =' ')
	print(np.array([r, 3 + k]))

	pivot = table[r][3 + k]
	print("elemento pivote : ", end =" ")
	print(Fraction(pivot).limit_denominator(100))
		

        # realiza operaciones de fila
    # divide la fila pivote con el elemento pivote
	table[r, 2:len(table[0])] = table[
			r, 2:len(table[0])] / pivot
			

    # hace operaciones de fila en otras filas
	i = 0
	while i<len(table):
		if i != r:
			table[i, 2:len(table[0])] = table[i,
				2:len(table[0])] - table[i][3 + k] * table[r, 2:len(table[0])]
		i += 1

	

    # asigna la nueva variable basica
	table[r][0] = k
	table[r][1] = c[k]
	
	print()
	print()
	itr+= 1
	

print()

print("*-*-*-*-*-*-*-*-*-*-*-*")
if unbounded == 1:
	print("No hay solucion")
	exit()
if alternate == 1:
	print("Solucion alterna")

print("tabla optima:")
print("B \tCB \tXB \ty1 \ty2 \ty3 \ty4")
for row in table:
	for el in row:
		print(Fraction(str(el)).limit_denominator(100), end ='\t')
	print()
print()
print("valor de z en la optimizacion: ", end =" ")

basis = []
i = 0
sum = 0
while i<len(table):
	sum += c[int(table[i][0])]*table[i][2]
	temp = "x"+str(int(table[i][0])+1)
	basis.append(temp)
	i+= 1

# si es un problema de minimizacion, hacer z negativo
if MIN == 1:
	print(-Fraction(str(sum)).limit_denominator(100))
else:
	print(Fraction(str(sum)).limit_denominator(100))
print("Basicos Finales: ", end =" ")
print(basis)

```
